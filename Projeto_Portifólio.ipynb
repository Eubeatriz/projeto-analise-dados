```python
import pandas as pd
import matplotlib.pyplot as plt
import random

# Gerar uma lista com 10 números aleatórios entre 1 e 100
dados = [random.randint(1, 100) for _ in range(10)]

# Exibir a lista gerada
print("Lista de dados", dados)

# Plotar os dados
plt.plot(dados, marker='o')  # marcador em cada ponto para melhor visualização
plt.title("Gráfico de Números Aleatórios")  # Correção: plt.title, não plt.plt.title
plt.xlabel("Índice")
plt.ylabel("Valor")
plt.grid(True)
plt.show()
```
